### Data Preprocessing

```python
iris_df.shape
```

```python
iris_df.isnull().sum()
```

```python
iris = iris_df.copy()
iris.drop(['Id','Species'],axis=1,inplace=True)
iris.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = iris.values[:,:]

Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
plt.style.use('fivethirtyeight')
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

You can clearly see why it is called 'The elbow method' from the above graph, the optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.

From this we choose the number of clusters as ** '3**'.

### Modeling

Let's apply K-means to the dataset and take look at cluster label 

```python
Num = 3
k_means = KMeans(init = "k-means++", n_clusters = Num, max_iter=400, n_init = 10,random_state=1)
y_kmeans = k_means.fit_predict(X)
labels = k_means.labels_
print(labels)
```

**Now we assign labels to the each row in the dataset.**

```python
iris["labels"] = labels
iris.head(5)
```

**We can check for the centroid values by averaging the features in each cluster.**

```python
iris.groupby('labels').mean()
```
